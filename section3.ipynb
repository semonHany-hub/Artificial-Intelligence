```python
import numpy 
arr=numpy.array([5,8,2])
print(type(arr), "\n")
print(arr.shape, "\n")
print(arr[0], arr[1], arr[2])
#numpy stands for numerical python. deals with matrices and linear algebra
```

```python
matrix=numpy.array([[5,7,2], [9,2,6]])
print(type(matrix), "\n")
print(matrix.shape, "\n")
print(matrix[0,1], matrix[1,2])
```

```python
onesArr=numpy.ones((1,2))
zerosArr=numpy.zeros((2,3))
unitMatrix=numpy.eye(2) #the identity matrix always square
constantMatrix=numpy.full((3,2), 7)
randomMatrix=numpy.random.random((2,3))
print("ones array: ", onesArr, "\n")
print("zeros array: ", zerosArr, "\n")
print("unit matrix: ", unitMatrix, "\n")
print("constant matrix: ", constantMatrix, "\n")
print("random matrix: ", randomMatrix) #matrix with random values from 0 to 1
```

```python
#indexing and slicing
matrix=numpy.array([[2,6,3], [8,3,5], [9,3,6]])
slice=matrix[0:2, 1:3]
slice[0,1]=77
print(matrix[0,2])
```

```python
# %pip install pandas
#pandas library stands for panel data. deals with series and data frames
import pandas as pd
s=pd.Series([4,7,2], index=['a', 'b', 'c']) #here we custom indexes of series, but the default indexes are 0,1,2,...
s
#pandas series is like a column 
```

```python
dict={"name":"semon", "age":20}
print(pd.Series(dict))
#dtype with value object represent the string
```

```python
data=[["semon", "Egypt", "Qena"], ["john", "british", "london"], ["aya", "suadi", "dubai"]]
"""we can use instead an object with keys:name, country and city, where the values types are lists
  or a list of object each object represent a row and has the keys:name, country and city"""
dataFrame=pd.DataFrame(data, columns=["name", "country", "city"])     
print(dataFrame)
#pandas data frame is like a table
```

```python
print(dataFrame["name"], "\n")
print(dataFrame[["country", "city"]], "\n")
#iloc used with default indexed elements
print(dataFrame.iloc[1], "\n") #return the second row
print(dataFrame.iloc[1, 0], "\n") #return the element in the second row and first column
dataFrame.index=[1,2,3] #custom rows indices
print(dataFrame.loc[2], "\n") #return the second custom indexed row
print(dataFrame[dataFrame["name"] =="semon"], "\n") #dataFrame[condition related to a column]
```

```python
print(dataFrame.head(1), "\n") #by default without any parameters it returns the first 5 rows
print(dataFrame.tail(2), "\n") #by default without any parameters it returns the last 5 rows
print(dataFrame.shape, "\n")
print(dataFrame.info(), "\n")
print(dataFrame.describe(), "\n")
```

```python
print(dataFrame.duplicated(subset="name"), "\n")
dataFrame.drop_duplicates(subset=["name"])
```

```python
dataFrame["age"]=[20, 17, 24]
dataFrame
```

```python
dataFrame=dataFrame.drop("age", axis=1) #return the data frame without the dropped column, but doesn't affect it in-place
dataFrame
# axis with value 0 affect the row, while with value 1 affect the column
```
