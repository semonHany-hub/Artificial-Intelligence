```python
#Dictionary
mydict = {1:'one' , 2:'two' , 3:'three'}
print("create dictionary with direct assignment to its structure {}: ", mydict)

mydict = dict({1:'one' , 2:'two' , 3:'three'})
print("Create dictionary using dict(): ",mydict)

print("dictionary keys: ", mydict.keys())
print("dictionary values: ", mydict.values())
print("dictionary items: ", mydict.items())

keys = {'a' , 'b' , 'c' , 'd'}
value = 10
mydict2=dict.fromkeys(keys, value)
print("create dictionary from keys and unified value: ", mydict2)
```

```python
#accessing items
mydict3 = {'Name':'Asif' , 'ID': 74123 , 'DOB': 1991 , 'job' :'Analyst'}
print("name: ", mydict3["Name"]) #access with bracket notation
print("job: ", mydict3.get("job")) #access with .get() method
```

```python
#Add, Remove & Change Dictionary Items
mydict3["DOB"]=2000 #update by accessing with bracket notation
mydict3.update({"job":"engineer", "Name":"semon"}) #update using the method .update(new values in a dictionary)
print("mydict3 after updating: ", mydict3)

mydict3["address"]="Egypt, Qena, Dishna"
print("mydict3 after adding address: ", mydict3)

mydict3.pop("ID")
mydict3.popitem()
print("mydict3 after removing ID and the last property: ", mydict3)

del[mydict3["Name"]]
print("mydict3 after deleting Name: ", mydict3)

mydict3.clear()
print("empty mydict3: ", mydict3)

del mydict3
mydict3
```
